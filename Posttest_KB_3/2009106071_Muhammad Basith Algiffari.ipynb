<h1>Import Library</h1>

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

<h1>Memanggil Dataset</h1>

```python
df = pd.read_csv('../Posttest_KB_2/BigBazaar.csv')
```

<h3>Menghilangkan Record yang memiliki value null</h3>

```python
df.dropna(inplace=True)
```

```python
df
```

### Mengecek banyak data sesuai dengan category

```python
value_categ = df['Category'].value_counts(ascending=True)
value_categ
```

## Penjelasan Grafik Di bawah ##
Pada bar dibawah ini memperlihatkan urutan jumlah barang yang dimiliki oleh setiap kategori pada kegiatan bazar ini dan dapat diambil kesimpulan bahwa:
 - Kategori Home Fashion memiliki jumlah barang terbanyak
 - Kategori Women memiliki jumlah barang terdikit
 - Data ini merupakan data yang telah dibersihkan dari record value null

```python
x = value_categ.index.tolist()
y = value_categ.astype(str)
f = plt.figure()
f.set_figwidth(20)
f.set_figheight(10)
plt.barh(x, y,color='Blue')
plt.title('Jumlah Kategori barang yang dijual', loc='center')
plt.xlabel('Banyak Barang', loc='center')
plt.ylabel('Kategori Barang', loc='center')
plt.show()
```

### Penjelasan Pie Chart

Pada Pie Chart Dibawah merupakan data 10 kategori yang memiliki jumlah barang terbanyak:
 - Bisa dilihat bahwa dari data pesebaran ini, Home Fashion memiliki bagian paling banyak dalam bazaar ini
 - Bisa dilihat juga kalau hal - hal yang berhubungan dengan stationery memiliki bagian paling sedikit

```python
data_pie =df['Category'].value_counts()[0:10].rename_axis('Category').reset_index(name='Id_count')
```

```python
plt.figure(figsize=(10,10))
plt.pie(data_pie.Id_count, labels=data_pie.Category, startangle=90, autopct='%.1f%%')
plt.title('Category')
plt.show()
```

## Penjelasan Scatter Plot

Scatter Plot dibawah adalah merupakan hubungan antara kategori dan pemberian discount dari setiap kategori:
 - Ada data outlier di kategori Home Appliance dan Electronics
 - Banyak barang yang mendapat discount dari range 0 - 2000

```python
sns.scatterplot(data = df, x ="DiscountedPrice", y = "Category")
plt.figure(figsize=(100,100))
plt.show()
```
