Melakukan import library yang akan digunakan

```python
import tensorflow as tf
import pandas as pd
from sklearn.model_selection import train_test_split
```

membaca dataset yang akan digunakan

```python
data = pd.read_csv('../Posttest_KB_5/diabetes.csv')
```

membagi data menjadi datax dan datay

```python
datax = data[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
datay = data[['Outcome']]
```

membagi dataset dengan ratio 80:20

```python
x_train, x_test, y_train, y_test = train_test_split(datax, datay, random_state=0, test_size=.2)
```

membagi dataset menjadi record dan label

```python
num_class = 2
y_train = tf.keras.utils.to_categorical(y_train, num_class)
y_test = tf.keras.utils.to_categorical(y_test, num_class)
```

membuat model deep learning menggunakan model sequential, dengan aktivasi relu dan menggunakan sigmoid karena kelas data hanya 2

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(32, activation='relu', input_dim=8),
    tf.keras.layers.Dense(num_class, activation='sigmoid')
])
model.summary()
```

melakukan configurasi pada model yang telah dipakai sebelumnya dengan optimizer adam dan lossnya binary crossentropy dan metrics akurasi

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(),
    loss=tf.keras.losses.binary_crossentropy,
    metrics=['accuracy']
)
```

melatih model sebanyak 200 kali dengan setiap langkah 10

```python
model.fit(
    x=x_train,
    y=y_train,
    validation_data=(x_test,y_test),
    epochs=200,
    steps_per_epoch=10,
)
```
